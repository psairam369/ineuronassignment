```python
def stat(data):
    mean=np.mean(data)
    median=np.median(data)
    mode=stats.mode(data)
    return mean,median,mode
```

```python
rateperhour=72
for4min=(rateperhour*4)/60
mu=for4min
def poss(k,mu):
    resval=poisson.pmf(k=k,mu=mu)
    return resval   
```

```python
#pictorial representation
res1=poisson.rvs(mu=mu,size=10000)
sns.distplot(res1, hist=True, kde=True,color="red")
```

# 7a.

```python
resval=poss(5,mu)
print(f"for the probability of 5 customers arriving in 4 minutes is {resval}")
```

# 7b.

```python
#not more than 3 
#k=0,1,2,3
#for k=0
resval0=poss(0,mu)
#for k=1
resval1=poss(1,mu)
#for k=2
resval2=poss(2,mu)
#for k=3
resval3=poss(3,mu)
result1=resval0+resval1+resval2+resval3
print(f"is the probability of not more than 3 customers arriving in 4 minutes is {result1}")

```

# 7c.

```python
#more than 3 customers
#more than 3 customers=1-probability of not more than 3 customers
nmt3customers=1-result1
print(f"is the probability of not more than 3 customers arriving in 4 minutes is {nmt3customers}")
```

```python
def pz(z):
    res=scipy.stats.norm.cdf(z)
    return res
```

a

```python
res=pz(1.26)
print(f"P(Z <1.26) is : {res}" )
#P(Z < −0.86)
res1=pz(-0.86)
print(f"P(Z < −0.86) is :{res1}")
#P(Z > −1.37)
res2=pz(-1.37)
print(f"P(Z > −1.37)is :{1-res2}")
#P(−1.25 < Z < 0.37)
#P(Z>-1.25)-P(Z<0.37)
res11=pz(-1.25)
resf11=res11 #P(Z<-1.25)
resf12=pz(0.37)  #P(Z<0.37)
fres=resf12-resf11  #P(Z<0.37)-P(Z<-1.25)
print(f"P(−1.25 < Z < 0.37) is : {fres}")
#P(Z ≤ −4.6)
res33=pz(-4.6)
print("P(Z ≤ −4.6) is  :"+str(res33))
```
